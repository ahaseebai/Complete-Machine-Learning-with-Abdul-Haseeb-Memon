# **Founders & Contributions - ML**

- **Founder:** Arthur Lee Samuel (He Coined the Term ML in 1959)<br>
- **Acheivement:** Built a self-learning checkers program at IBM that improved by experience<br>
- **Significance:** Established the core idea of learning from data rather than fixed rules<br>
- **Theoretical Foundations:** Alan Turing (Father of Ai)
- **Contributions of Alan Turing**: Alan Turing provided the theoretical groundwork for machines that could process information and potentially "think". In his influential 1950 paper, "Computing Machinery and Intelligence," he introduced the "Turing Test" as a criterion for machine intelligence, his research gives idea into thinking machines long before the technology existed. <br>

- **Tom M. Mitchell (1997):** *provided the most cited formal definition of Machine Learning:* A computer program learns from experience E with respect to tasks T and performance measure P if its performance at tasks T, as measured by P, improves with experience E.



# **Difference b/w Programming & Machine Learning** ðŸŽ­

---

  | Aspect             | Traditional Programming              | Machine Learning                             |
  | ------------------ | ------------------------------------ | -------------------------------------------- |
  | **Core Idea**      | Human writes explicit rules          | System learns rules from data                |
  | **Input**          | Data + Program (rules)               | Data + Expected Output (labels or feedback)  |
  | **Process**        | Logic is hard-coded                  | Patterns are learned automatically           |
  | **Output**         | Deterministic result                 | Predictive or probabilistic result           |
  | **Rule Creation**  | Manual (by programmer)               | Automatic (by algorithm)                     |
  | **Adaptability**   | Does not improve unless reprogrammed | Improves with more experience (data)         |
  | **Error Handling** | Fixed logic â†’ fixed mistakes         | Errors reduce as learning continues          |
  | **Scalability**    | Poor for complex patterns            | Excellent for complex, high-dimensional data |
  | **Human Role**     | Writes exact instructions            | Designs task, data, and evaluation           |
  | **Best Use Case**  | Clear, rule-based problems           | Pattern recognition, prediction, uncertainty |
