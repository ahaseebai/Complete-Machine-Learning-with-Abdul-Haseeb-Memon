# **Data Foundations** ‚åõ
  **DATA TERMS**
  - **Data:** Raw facts. No meaning by itself.
  - **Information:** Data that is processed and has meaning.
  - **Knowledge:** Understanding patterns and rules from information.
  - **Learning:** The process of moving upward (Data ‚Üí Wisdom). its affect
  - **Training:** its act that produces any affect
  - **Testing:** its for new data to test either its learnt or not
  - **Wisdom:** Using knowledge to make correct decisions.

  **DATA STORAGE & STRUCTURE**
  - **Dataset:** A collection of related data used for analysis or training models.

  - **Data Files:** Physical files that store data (CSV, Excel, JSON, TXT).

  - **Data Frames:** Table-like data structure (rows + columns) used in analysis (e.g., Pandas DataFrame).

  - **Database:** An organized system to store, manage, and query large data efficiently.

  **DATA TYPES**
  - **Structured Data:** Well-organized, tabular data (rows/columns) ‚Üí easy to query (SQL).

  - **Unstructured Data:** No fixed format (text, images, videos, audio).

  - **Semi-Structured Data:** Partially organized (JSON, XML, logs).

  **DATA COMPONENTS**
  - **Feature / Attribute / Column:** An input variable (e.g., age, price).
  - **Label / Target:** The output you want to predict (e.g., pass/fail).
  - **Observation / Record / Row:** One complete data entry (one student, one book).
  - **Index:** A unique identifier for rows.
  - **Metadata:** Data about data (file size, author, date created).

  **DATA PROCESSING**
  - **Data Cleaning / Preprocessing:** Fixing missing values, errors, noise.
  - **Data Wrangling / Munging:** Transforming raw data into usable form.
  - **Data Manipulation:** Sorting, filtering, aggregating data.
  - **Data Extraction:** Pulling data from files, databases, APIs.

  **DATA FLOW & PIPELINES**
  - **Data Pipeline:** Automated flow of data from source to destination.
  - **ETL (Extract, Transform, Load):** Extract data ‚Üí clean/transform ‚Üí store in database.
  - **Batch Processing:** Process data in chunks (daily, weekly jobs).
  - **Real-Time / Stream Processing:** Process data instantly as it arrives (live data).

  **DATABASE TECHNOLOGIES**
  - **SQL:** Query language for structured relational databases.
  - **NoSQL Database:** Databases for flexible, large-scale, unstructured data.

  **DATA DOMAINS**
  - **Data Analysis:** Finding insights, patterns, trends from data.
  - **Data Science:** Using data + statistics + ML to solve problems.
  - **Data Engineering:** Building systems to collect, store, and move data.

  **MODERN DATA CONCEPTS**
  - **Big Data:** Extremely large, fast, complex data that traditional systems can‚Äôt handle.
  - **API:** Interface that allows systems to exchange data.
  - **Dimensionality:** Number of features in a dataset.
  - **Data Modeling:** Designing how data is structured and related.

  **GOVERNANCE & VISUALS**
  - **Data Visualization:** Representing data using charts and graphs.
  - **Data Governance:** Rules and policies for data quality and control.
  - **Data Ethics / Privacy:** Using data responsibly and protecting user data.

#**Feature Engineering** ‚ö°
  - **Categorical Encoding**
    - **Label Encoding:** Assign numeric labels to categories
    - **One Hot EnCoding:** Convert categories into binary vectors.
    - **Target (mean) EnCoding:** Convert categories into binary vectors.

    - **Binary EnCoding:** Convert categories into binary vectors.

  - **Feature Scalling**
    - **Normalizing:** Scaling values to a [0,1] range.
    - **standardization:** Scaling values to have mean 0 and standard deviation 1.
  - **Feature Selection**
    - **Correlation:** Measures the strength and direction of the relationship between two variables.

    - **Forward Elimination:** Iteratively adds features to a model based on statistical significance.
    - **Backward Elimination:** Iteratively removes the least significant features from a model.
    - **Univariate Selection:** Selects the best features based on statistical tests with the target variable.
    - **Random Forest Importance:** Determines feature importance based on how much each feature improves prediction in a Random Forest model.
    - **Feature Selection with Decision Trees:** Uses decision tree splits to identify and keep the most influential features.
