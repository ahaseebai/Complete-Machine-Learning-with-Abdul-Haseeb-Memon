## **Encoding: conversion of Categorical Data into Numerical Data**

````python
import pandas as pd
Dataset=pd.read_csv("diagnosis.csv")
Dataset
````

````python
Dataset=Dataset.fillna(method="bfill")
````

# **Label Encoding**

````python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
label=le.fit_transform(Dataset["Diagnosis "])
Dataset["Label Encoded"]=label
````

# **Ordinal Encoding**

````python
import pandas as pd
ord_data={
    "small" : 0,
    "medium" : 1,
    "large": 2,
    "xl": 3
}
data=pd.DataFrame({
    "Size": ["small","large","small","large","xl","medium","medium","xl"]
})
new=data["Size"].map(ord_data)
data["Ordinal_Encoded"]=new
````

````python
data
````

````python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories=[["small", "medium", "large", "xl"]])
encoded = oe.fit_transform(data[["Size"]])
data["Ordinal_Encoded"] = encoded
data
````
