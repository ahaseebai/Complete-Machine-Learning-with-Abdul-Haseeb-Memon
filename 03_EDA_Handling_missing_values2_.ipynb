# **Handle Categorical Columns**

````python
import pandas as pd
Dataset=pd.read_csv("diagnosis.csv")
Dataset
````


# **Checking Missing Values Column Wise**

````python
Dataset.isnull().sum()
````

````python
Dataset.info()
````

# **Handle Missing Value by Mode**



```
Dataset["Column"].fillna(Dataset["Column"].mode()[0],inplace=True)
```



# **Handle Missing Value by Forward Filling**

````python
Dataset.fillna(method="ffill")
````

# **Handle Missing Value by Backward Filling**

````python
Dataset.fillna(method="bfill")
````

# **Handle by for Loop**

````python
for i in Dataset.select_dtypes(include="object").columns:
  a=Dataset[i].fillna(method="bfill")
  print(a)
````

# **Handle Missing Values by Scikit Learn (Simple Imputer)**

````python
from sklearn.impute import SimpleImputer

# Imputer for categorical column(s)
si = SimpleImputer(strategy="most_frequent")

# Select only object (categorical) columns
cat_cols = Dataset.select_dtypes(include="object").columns.tolist()

# Fit and transform categorical columns
array = si.fit_transform(Dataset[cat_cols])

# Convert back to DataFrame with same column names
new_missing = pd.DataFrame(array, columns=cat_cols)
````

````python
new_missing.isnull().sum()
````

````python
Dataset["Diagnosis"]=new_missing
Dataset
````
