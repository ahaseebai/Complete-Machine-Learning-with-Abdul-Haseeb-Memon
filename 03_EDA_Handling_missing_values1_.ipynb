# **Handling a Missing Values**
- Numerical data (Mean / Median)
- Categorical Data (Mode / orward Filling / Backward Filling)

```python
Dataset["Age (X)"].fillna(Dataset["Age (X)"].median())
```

```python
Dataset.isnull().sum()
```

```python
Dataset["Blood Pressure (Y)"].fillna(Dataset["Blood Pressure (Y)"].mean())
```

# **Checking Missing Values: as it has been handled Successfully**

```python
Dataset.isnull().sum()
```

# **Handle Missing Values by Scikit Learn (Simple Imputer)**


```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(strategy="mean")
array=si.fit_transform(Dataset[["Age (X)","Blood Pressure (Y)"]])
new_missing = pd.DataFrame(
    array,
    columns=["Age (X)", "Blood Pressure (Y)"]
)
```

```python
new_missing.isnull().sum()
```
