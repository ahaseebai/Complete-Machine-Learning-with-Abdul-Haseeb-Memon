# **Feature scaling**

````python
import pandas as pd
ds=pd.read_csv("record.csv")
ds.head(3)
````

````python
ds.isnull().sum()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(ds)
plt.show()
````

# **Feature Scaling**

Standardization: mean = 0, varience =1
- `formula = (xi - mean)/std`

````python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(ds[["ID"]])
ds["ID_Scaled"] = (ss.transform(ds[["ID"]]))
````

````python
ds.head(5)
````

````python
sns.displot(ds["ID_Scaled"])
plt.show()
````

# **Comparing previous graph with new one after Standardization**

````python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(ds["ID"])
plt.subplot(1,2,2)
plt.title("After")
sns.distplot(ds["ID_Scaled"])
plt.show()
````

# **Normalization:**


```
- values shifted b/w 0 - 1
- Formula= (xi - min(x))/(max(x) - min(x))


```



````python
from sklearn.preprocessing import MinMaxScaler
minmax_scale = MinMaxScaler()
minmax_scale.fit(ds[["Score"]])
ds["Score_Scaled"] = (minmax_scale.transform(ds[["Score"]]))
````

````python
ds.head(5)
````

````python
sns.displot(ds["Score_Scaled"])
plt.show()
````

# **Comparing previous graph with New one after Normalization**

````python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(ds["Score"])
plt.subplot(1,2,2)
plt.title("After")
sns.distplot(ds["Score_Scaled"])
plt.show()
````
