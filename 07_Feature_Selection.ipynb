```python
import pandas as pd
ds=pd.read_csv("f_selection.csv")
ds.head(3)
```

```python
ds.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(ds)
plt.show()
```

# **Feature Selection**
# **Forward Elimination**

```python
from mlxtend.feature_selection import SequentialFeatureSelector
X=ds.iloc[:,:-1]
y=ds['Feature_5']

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
fs=SequentialFeatureSelector(lr,k_features=4,forward=True)
fs.fit(X,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_*100
```

# **Backward Elimination**

```python
from mlxtend.feature_selection import SequentialFeatureSelector
X=ds.iloc[:,:-1]
y=ds['Feature_5']

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
fs=SequentialFeatureSelector(lr,k_features=4,forward=False)
fs.fit(X,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_*100
```
